```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# plotting templates
sns.set(context="talk", style="ticks") 
pal = sns.color_palette("tab10")
```

#### Create a dataframe to hold the results to be plotted

```python
cohorts = ["BRCA", "CCRCC", "COAD", "GBM", "HGSC", "HNSCC", "LSCC", "LUAD", "PDAC", "UCEC"]
savage2024_DE_genes=[np.nan, 2126, 1986, np.nan, 1376, 2306, 2855, 2914, 1431, 999]
n_overlap=[np.nan, 2020, 1954, np.nan, 1322, 2118, 2811, 2851, 1267, 802]

plotter = pd.DataFrame(columns=["cohorts", "savage2024", "n_overlap"])
plotter["cohorts"] = cohorts
plotter["savage2024"] = savage2024_DE_genes
plotter["n_overlap"] = n_overlap
```

#### Get the percentage of overlap

```python
overlap_frac = plotter["n_overlap"] / plotter["savage2024"]
plotter["overlap_frac"] = overlap_frac
plotter["savage_exclusive"] = 1 - plotter["overlap_frac"]
```

#### Plot, with the legend

```python
cust_palette = ["#ff7f0e", "#1f77b4"]
labels = ["overlap", "Savage 2024 \nexclusive"]

fig, ax = plt.subplots(2, 4, figsize=(14, 6))

# CCRCC
# patches, text1, text2 = \
#     ax[0,0].pie([plotter["overlap_frac"][1], plotter["savage_exclusive"][1]], 
#             colors=cust_palette, autopct='%.0f%%')
# ax[0,0].legend(patches, labels, loc="upper left", bbox_to_anchor=(0,0))
# ax[0,0].set_title("CCRCC", size=16, pad=1)

ax[0,0].pie([plotter["overlap_frac"][1], plotter["savage_exclusive"][1]], 
            colors=cust_palette, autopct='%.0f%%')
ax[0,0].set_title("CCRCC", size=16, pad=1)

# COAD
ax[0,1].pie([plotter["overlap_frac"][2], plotter["savage_exclusive"][2]], 
            colors=cust_palette, autopct='%.0f%%')
ax[0,1].set_title("COAD", size=16, pad=1)

# HGSC
ax[0,2].pie([plotter["overlap_frac"][4], plotter["savage_exclusive"][4]], 
            colors=cust_palette, autopct='%.0f%%')
ax[0,2].set_title("HGSC", size=16, pad=1)

# HNSCC
# ax[0,3].pie([plotter["overlap_frac"][5], plotter["savage_exclusive"][5]], 
#             colors=cust_palette, autopct='%.0f%%')
# ax[0,3].set_title("HNSCC", size=16, pad=1)

patches, text1, text2 = \
    ax[0,3].pie([plotter["overlap_frac"][5], plotter["savage_exclusive"][5]], 
            colors=cust_palette, autopct='%.0f%%')
ax[0,3].legend(patches, labels, loc="upper left", 
               bbox_to_anchor=(1.05,0.1), edgecolor="k", prop={"size":16})
ax[0,3].set_title("HNSCC", size=16, pad=1)

# LSCC
ax[1,0].pie([plotter["overlap_frac"][6], plotter["savage_exclusive"][6]], 
            colors=cust_palette, autopct='%.0f%%')
ax[1,0].set_title("LSCC", size=16, pad=1)

# LUAD
ax[1,1].pie([plotter["overlap_frac"][7], plotter["savage_exclusive"][7]], 
            colors=cust_palette, autopct='%.0f%%')
ax[1,1].set_title("LUAD", size=16, pad=1)

# PDAC
ax[1,2].pie([plotter["overlap_frac"][8], plotter["savage_exclusive"][8]], 
            colors=cust_palette, autopct='%.0f%%')
ax[1,2].set_title("PDAC", size=16, pad=1)

# UCEC
ax[1,3].pie([plotter["overlap_frac"][9], plotter["savage_exclusive"][9]], 
            colors=cust_palette, autopct='%.0f%%')
ax[1,3].set_title("UCEC", size=16, pad=1)

fig.show()
#plt.savefig("../figures/DE-comparison-w-paper-piecharts.pdf", bbox_inches="tight")
```

#### Try stacked barplot instead of pie charts

```python
plt.figure(figsize=(6,5))

plt.bar(
    plotter["cohorts"], plotter["overlap_frac"], 
    color="#1f77b4", label="Overlap",
)
plt.bar(
    plotter["cohorts"], plotter["savage_exclusive"], 
    bottom=plotter["overlap_frac"], color="#ff7f0e", label="Savage Excusive",
)

plt.xlabel("", size=20, labelpad=14)
plt.ylabel("Fraction", size=20, labelpad=14)

plt.xticks(rotation=90)
plt.legend(title="", edgecolor="k", prop={'size': 16}, 
           loc="upper center", bbox_to_anchor=(0.5, 1.25))

#plt.show()
#plt.savefig("../figures/DE-comparison-w-paper-barplot.pdf", bbox_inches="tight")
```
